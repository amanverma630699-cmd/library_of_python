```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(1,10,100)
y=np.sin(x)
y1=np.cos(x)
y2=np.cos(x)*np.sin(x)

plt.figure(figsize=(10,6))
plt.plot(x,y,label='sin (x)')
plt.plot(x,y1, label='cos(x)')
plt.plot(x,y2 ,label='sin(x)*cos(x)')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('try')
plt.legend()
plt.grid(True ,alpha=0.3)
plt.show()

```

```python
x=np.linspace(0,10,20)
plt.figure(figsize=(12,6))
# plt.figure(figsize=(12, 6))

plt.plot(x,x+1,'b-',label='solid',linewidth=2)
plt.plot(x,x+2,'r-.',label='dash dot',linewidth=2)
plt.plot(x,x+3,'g--',label='dash',linewidth=2)
plt.plot(x,x+3,'g:',label='dot',linewidth=2)
plt.plot(x,x+4,'mo-',label='cicle',markersize=6)
plt.plot(x,x+5,'sc-',label='square',markersize=6)
plt.plot(x,x+5,'^k-',label='triangle',markersize=6)
plt.xlabel("x aixe")
plt.ylabel("y aixe")
plt.legend()
plt.grid(True,alpha=0.3)
plt.show()
```

```python
np.random.seed(415)
x=np.random.randn(100)
y=np.random.randn(100)
colors = np.random.rand(100)  # Random colors
sizes = 1000 * np.random.rand(100)
plt.figure(figsize=(10,6))
plt.scatter(x,y ,c=colors,s=sizes,alpha=0.3,cmap='Set1')
plt.colorbar(label='colarint')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
